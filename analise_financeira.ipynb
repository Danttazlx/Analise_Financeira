```python

# Carregar os Dados 

import pandas as pd
import numpy as np


df_base = pd.read_csv('clientes_financeiro.csv', sep=',' ,low_memory=False)

df_base
```

```python
# inspecionando os Dados 

display(df_base.shape)    # 20,9
display(df_base.dtypes) 
#display(df_base.describe())
```

```python
df_analise = df_base
```

```python
# padronização

df_analise['nome'] = df_analise['nome'].str.strip().str.title() 
```

```python
df_analise
```

```python
# Tratamento dos Dados 


df_analise['limite_disponivel'] = df_analise['gasto_mensal'] / df_analise['limite_credito']
df_analise
```

```python
# trocando o nome da coluna 

df_analise = df_analise.rename(columns={'limite_disponivel' : 'taxa_utilizacao'})
df_analise
```

```python
# análise pela taxa de utilizacao (limite credito)

df_analise['taxa_utilizacao'] = df_analise['taxa_utilizacao']

df_risco = df_analise[df_analise['taxa_utilizacao'] > 0.8]
display(df_risco[['nome', 'cidade', 'status_pagamento', 'taxa_utilizacao']])
```

```python
# colocando em ordem decrescente pela utilização 

display(df_risco[['nome', 'cidade', 'status_pagamento', 'taxa_utilizacao']].sort_values(by='taxa_utilizacao' , ascending= False))
```

```python
# analise por inadiplente 



df_inadimplente = df_analise[df_analise['status_pagamento'] == 'Inadimplente']

display(df_inadimplente[['nome', 'idade', 'renda_mensal','cidade']])
display(df_inadimplente[['idade', 'renda_mensal']].mean())

```

```python
df_analise.to_csv('clientes_financeiro_analisado.csv', index=False)
df_risco.to_csv('clientes_risco_e_inadimplencia.csv', index=False)
print('Arquivo Salvo com sucesso')
```
